```python
import numpy as np
import pandas as pd
```

```python
with open("/kaggle/input/projectgutenberg/book.txt","r") as file:
    content = file.read();
    content = content.lower();
    print(content);
```

```python
sentences = content.split('\n');
print(sentences);
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(oov_token = '<UNK>');
```

```python
tokenizer.fit_on_texts(sentences);
```

# tokenizer.word_index a dictionary that maps each unique word in your text corpus to a unique integer, based on frequency (most frequent word gets the smallest index like 1, then 2, and so on).

```python
print(tokenizer.word_index);
```

```python
vocab_size = len(tokenizer.word_index) + 1;
print(vocab_size);
```

```python
sequences = tokenizer.texts_to_sequences(sentences);
print(sequences);
```

```python
input_sequences = [];
for sequence in sequences:
    for i in range(1,len(sequence)):
        n_gram_sequence = sequence[:i+1];
        input_sequences.append(n_gram_sequence);
```

```python
print(input_sequences[0],input_sequences[1]);
```

# all these sequences are of different lengths
# To train an RNN model on this , we need all sequences to be of the same length
# to do this , we can pad the sequences
# Padding means to find the maximum sequence length in the list of sequences and pad all the shorter sequences 

```python
max_seq_len = max([len(seq) for seq in input_sequences]);
print(max_seq_len);
```

```python
input_sequences = pad_sequences(input_sequences,maxlen = max_seq_len);
```

```python
print(input_sequences[0],input_sequences[1]);
```

```python
sequences = np.array(input_sequences);
x = sequences[:,:-1];
labels = sequences[:,-1];
```

```python
y = tf.keras.utils.to_categorical(labels);
print(x.shape,y.shape);
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout,Embedding,Bidirectional,LSTM,LayerNormalization
from tensorflow.keras.optimizers import Adam
model = Sequential();
```

# Embedding Layer represents the word in n dimensional space
# Bidirectional means it is taking the context from both directions

```python
model.add(Embedding(vocab_size,100,input_length = max_seq_len - 1));
model.add(Bidirectional(LSTM(256,return_sequences = True)));
model.add(Dropout(0.2));
model.add(LSTM(256));
model.add(LayerNormalization())
model.add(Dense(512,activation = "relu"));
model.add(Dropout(0.2));
model.add(Dense(vocab_size,activation = "softmax"));
```

```python
adam = Adam(learning_rate = 0.01);
model.compile(optimizer = adam,loss = "categorical_crossentropy",
metrics = ["accuracy",tf.keras.metrics.Precision(),tf.keras.metrics.Recall(),tf.keras.metrics.AUC()]);
```

```python
from tensorflow.keras.callbacks import EarlyStopping
earlystopping = EarlyStopping(monitor = "accuracy",patience = 10,min_delta = 0.005,restore_best_weights = True)
```

```python
history = model.fit(x,y,epochs = 100,batch_size = 512,callbacks = [earlystopping]);
```

```python
print(model.summary());
```

```python
seed_text = "I could not help Laughing at the ease with which he explained his process of deduction";
next_words = 100;
for num in range(next_words):
    sequence = tokenizer.texts_to_sequences([seed_text]);
    sequence = pad_sequences(sequence,maxlen = max_seq_len - 1);
    predicted = model.predict(sequence,verbose = 0);
    predicted = predicted.argmax(axis = -1)[0];
    output_word = '';
    for word ,index in tokenizer.word_index.items():
        if(index == predicted):
            output_word = word;
            break;
    seed_text += ' ' + output_word;
print(seed_text);
```

```python
from matplotlib import pyplot as plt
fig,ax = plt.subplots(2,1,figsize = (18,10));
ax[0].plot(history.history["loss"],color = "b",label = "Training Loss");
ax[0].set_xlabel("Epochs");
ax[0].set_ylabel("Loss");
ax[0].set_title("Loss Curves");
ax[0].legend(loc = "best",shadow = True);
ax[1].plot(history.history["accuracy"],color = "b",label = "Training Accuracy");
ax[1].set_xlabel("Epochs");
ax[1].set_ylabel("Accuracy");
ax[1].set_title("Accuracy Curves");
ax[1].legend(loc = "best",shadow = True);
plt.tight_layout();
```
